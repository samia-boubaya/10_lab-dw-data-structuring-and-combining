### OPTION ONE: combine the three datasets and then apply the cleaning function to the new combined dataset

```python
import pandas as pd

# read the three csv files into seperate dataframes
df1 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file1.csv')
df2 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file2.csv')
df3 = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/file3.csv')

# display the 1st dataframes
print("DataFrame 1:")
print("Shape of df1:", df1.shape)
display(df1.head())
print(*80*"-")

# display the 2nd dataframes
print("DataFrame 2:")
print("Shape of df2:", df2.shape)
display(df2.head())
print(*80*"-")

# display the 3rd dataframes
print("DataFrame 3:")
print("Shape of df3:", df3.shape)
display(df3.head())
print(*80*"-")

# standardize column names in all three dataframes to the same column names
# 'State' to 'ST' in df3
df3 = df3.rename(columns={'State': 'ST'})
# 'Gender' to 'GENDER' in df3
df3 = df3.rename(columns={'Gender': 'GENDER'})

    
# ------------------------------------------------------------------------------------------------------------------------------------------------
# concatenate the three csv files into one dataframe
df_main = pd.concat([df1, df2, df3], ignore_index=True)

# display the main dataframe
print("Main DataFrame after concatenation:")
print("Shape of df_main:", df_main.shape)
display(df_main.head())

# save the main dataframe to a csv file
df_main.to_csv('combined_00_concat.csv', index=False)
```

```python
import pandas as pd
# Load the standardized dataset into a new DataFrame
df = pd.read_csv("combined_01_standard.csv")

# 01 - gender ---------------------------------------------------------------------------
print("Before: 'gender' unique values:", df['gender'].unique())
# mapping gender values
gender_mapping = {
    "female": "F",
    "Femal": "F",    
    "Male": "M"
}
# replace values in gender
df['gender'] = df['gender'].replace(gender_mapping)
# Check GENDER unique values after mapping
print("After: 'gender' unique values:", df['gender'].unique())
print(*70*"-")

# 02 - state ---------------------------------------------------------------------------
print("Before: 'state' unique values:", df['state'].unique())
# mapping gender values
state_mapping = {
    "WA": "Washington",
    "AZ": "Arizona",    
    "Cali": "California"
}
# replace values in state
df['state'] = df['state'].replace(state_mapping)
# Check STATE unique values after mapping
print("After: 'state' unique values:", df['state'].unique())
print(*70*"-")

# 03 - education ---------------------------------------------------------------------------
print("Before: 'education' unique values:", df['education'].unique())
# replace Bachelors with Bachelor in education_level
df['education'] = df['education'].replace('Bachelors', 'Bachelor')
# Check education unique values after mapping
print("After: 'education' unique values:", df['education'].unique())
print(*70*"-")

# 04 - customer_lifetime_value ---------------------------------------------------------------------------
print("Before: 'customer_lifetime_value' unique values:", df['customer_lifetime_value'].unique())
# Remove '%' and convert to float
df['customer_lifetime_value'] = df['customer_lifetime_value'].str.replace('%', '', regex=False).astype(float)
# Check education_level unique values after mapping
print("After: 'customer_lifetime_value' unique values:", df['customer_lifetime_value'].unique())
print(*70*"-")

# 05 - vehicle_class --------------------------------------------------------------------------
print("Before: 'vehicle_class' unique values:", df['vehicle_class'].unique())
# mapping vehicle_class values
vehicle_mapping = {
    "Sports Car": "Luxury",
    "Luxury SUV": "Luxury",
    "Luxury Car": "Luxury"
}
# replace values in education_level
df['vehicle_class'] = df['vehicle_class'].replace(vehicle_mapping)
# Check unique values after mapping
print("After: 'vehicle_class' unique values:", df['vehicle_class'].unique())
print(*70*"-")

# 06 - number_open_complaints --------------------------------------------------------------------------
print("Before: 'number_open_complaints' unique values:", df['number_open_complaints'].unique())
# Extract the number of open complaints from the string and convert to numeric
df['number_open_complaints'] = pd.to_numeric(df['number_open_complaints'].str.split('/', expand=True)[1],errors='coerce')
# Check unique values after mapping
print("After: 'number_open_complaints' unique values:", df['number_open_complaints'].unique())
print(*70*"-")

# Display the dataset after dealing with invalid values ----------------------------------------------------------------------
print("Dataset after dealing with invalid values:")
display(df.head())

# Save step 02 deal with invalid values dataset into a new CSV file ------------------------------
df.to_csv("combined_02_invalid-values.csv", index=False)
```

```python
import pandas as pd
# read csv
df = pd.read_csv("combined_03_data-formatting.csv")

# Check for null values
null_counts = df.isnull().sum()
print(f"Total null count in the Dataframe = {null_counts.sum()}\n")

# Identify how many null values each column has. 
print("Columns with null values:")
display(null_counts[null_counts > 0])

print("-" * 70)

# show null values in each column
print("Null values in each column:")
# Loop through each column and print the number of nulls in each column
for col in df.columns:
    null_count = df[col].isnull().sum()
    print(f"Column '{col}'‚û°Ô∏è[{null_count}] null values")
```

```python
# show unique values of each column
print("\nUnique values in each column:")
for col in df.columns:
    unique_vals = df[col].unique()
    print(f"{col}")
    print(f"Unique values ({len(unique_vals)}): {unique_vals}\n")
```

```python
# null values in the dataset before dealing with null values
null_counts_before = df.isnull().sum()


# null values in education column BEFORE
print("BEFORE, null values in education column:", df['education'].isnull().sum())
# fill null values in education column with 'Not Specified'
df['education'] = df['education'].fillna('Not Specified')
# null values in education column AFTER
print("AFTER, null values in education column:", df['education'].isnull().sum())
print("-" * 70)

# null values in customer_lifetime_value BEFORE
print("BEFORE, null values in customer_lifetime_value column:", df['customer_lifetime_value'].isnull().sum())
# fill null values in customer_lifetime_value column with median value of the column
df['customer_lifetime_value'] = df['customer_lifetime_value'].fillna(df['customer_lifetime_value'].median())
# null values in customer_lifetime_value column AFTER
print("AFTER, null values in customer_lifetime_value column:", df['customer_lifetime_value'].isnull().sum()) 
print("-" * 70)

# null values in gender column
print("null values in gender column:", df['gender'].isnull().sum())
# fill null values in gender column with 'U' for unknown
df['gender'] = df['gender'].fillna('U')
# null values in gender column AFTER
print("AFTER, null values in gender column:", df['gender'].isnull().sum())
print("-" * 70)

# null values in state column
print("null values in state column:", df['state'].isnull().sum())
# fill null values in state column with 'Not Specified'
df['state'] = df['state'].fillna('Not Specified')
# null values in state column AFTER
print("AFTER, null values in state column:", df['state'].isnull().sum())
print("-" * 70)

# null values in number_open_complaints column
print("null values in number_open_complaints column:", df['number_open_complaints'].isnull().sum())
# fill number_open_complaints null values with 0
df['number_open_complaints'] = df['number_open_complaints'].fillna(0)
# null values in number_open_complaints column AFTER
print("AFTER, null values in number_open_complaints column:", df['number_open_complaints'].isnull().sum())
print("-" * 70)

# null values in vehicle_class column
print("null values in vehicle_class column:", df['vehicle_class'].isnull().sum())
# fill null values in vehicle_class column with 'Not Specified'
df['vehicle_class'] = df['vehicle_class'].fillna('Not Specified')
# null values in vehicle_class column AFTER
print("AFTER, null values in vehicle_class column:", df['vehicle_class'].isnull().sum())
print("-" * 70)

# null values in customer_id column
print("null values in customer_id column:", df['customer_id'].isnull().sum())
# drop rows with null values in customer_id column
df = df.dropna(subset=['customer_id'])
# null values in customer_id column AFTER
print("AFTER, null values in customer_id column:", df['customer_id'].isnull().sum())
print("-" * 70)

# after dealing with null values in the dataset, check for null values again
null_counts_after = df.isnull().sum()
print("Null values in each column after dealing with null values:")
display(null_counts_after)

# save in a new csv file the dataset after dealing with null values
df.to_csv("combined_04_no-nulls.csv", index=False)

# display the dataset after dealing with null values ----------------------------------------------------------------------
print("Dataset after dealing with null values:")
print("Dataset shape:", df.shape)
display(df.head())
```

```python
# Define logical order for education
education_order = ['High School', 'College', 'Master', 'Doctor', 'PhD']
df['education'] = pd.Categorical(df['education'], categories=education_order, ordered=True)

# Nested pivot table: education as columns, gender as rows
pivot_gender_education = df.pivot_table(
    index='gender',
    columns='education',
    values='customer_lifetime_value',
    aggfunc='mean'
).round(2)  # round values

# Optional: reset columns if you want a flat table
pivot_gender_education = pivot_gender_education.reset_index()

pivot_gender_education
```
